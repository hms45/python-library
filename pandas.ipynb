<span style="background-color:yellow">람다식에서 여러 개의 값을 입력 인자로 할때 map 함수</span>

```python
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x : len(x))
titanic_df.head()
```

```python
titanic_df['Age_range'] = titanic_df['Age'].apply(lambda x: 'Child' if x > 19 else 'Adult')
titanic_df[['Age', 'Age_range']].head(10)
```

<span style="background-color:yellow">lambda식에 if조건문 가능</span><br>
<span style="background-color:yellow">lambda x : 'value' if (조건) else 'else value'</span>

```python
titanic_df['Age'] = titanic_df['Age'].astype(int)

def age_band(age):
    age_band = ''
    if age < 20: age_band = '20살 미만'
    elif age < 30: age_band = '20대'
    elif age < 40: age_band = '30대'
    elif age < 50: age_band = '40대'
    elif age < 60: age_band = '50대'
    elif age < 70: age_band = '60대'
    else: age_band = '70대 이상'

    return age_band

titanic_df['Age_band'] = titanic_df['Age'].apply(lambda x : age_band(x))
titanic_df[['Age','Age_band']].head(10)
```
